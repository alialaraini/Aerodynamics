```python
import math 
import numpy as np
from matplotlib import pyplot 
```

```python
x_start, x_end = -2.0, 2.0
y_start, y_end = -1.0, 1.0
N = 50 #Number of divisions we give to 
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)

print ('x=',x)
print ('y=',y)
# just created the array , mesh grid function, 
X,Y= numpy.meshgrid(x,y)
```

```python
%matplotlib inline 

width = 10
height = (y_end-y_start)/(x_end-x_start)*width #with this you can control the size of the figure 
pyplot.figure(figsize =(width,height))
pyplot.xlabel('x', fontsize=16)
pyplot.xlabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start,y_end)

pyplot.scatter(X,Y, s=5, color='red', marker = 'o')
```

source flow

```python

#strength of source = ss
ss = 5.0
x_source, y_source = -1.0,0

u_source = ss/(2*math.pi)*(X-x_source)/(((X-x_source)**2)+((Y-y_source)**2))
v_source = ss/(2*math.pi)*(Y-y_source)/(((X-x_source)**2)+((Y-y_source)**2))
```

```python
width = 10.0
height = (y_end-y_start)/(x_end-x_start)*width #with this you can control the size of the figure 
pyplot.figure(figsize =(width,height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,u_source, v_source, density=1.5, linewidth=1, arrowsize=2, arrowstyle = '->' )
pyplot.scatter(x_source,y_source,color='red', s=80, marker='o')
```

sink flow 

```python
#strengh_sink = Ss
Ss = -5
x_sink, y_sink =1.0,0.0
u_sink = Ss/(2*math.pi)*(X-x_sink)/(((X-x_sink)**2)+((Y-y_sink)**2))
v_sink = Ss/(2*math.pi)*(Y-y_sink)/(((X-x_sink)**2)+((Y-y_sink)**2))

width = 10
height = (y_end-y_start)/(x_end-x_start)*width #with this you can control the size of the figure 
pyplot.figure(figsize =(width,height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start,y_end)

pyplot.streamplot(X,Y,u_sink, v_sink, density=2, linewidth=1, arrowsize=2, arrowstyle = '->' )
pyplot.scatter(x_sink,y_sink,color='red', s=80, marker='o')
```

recaulate the flow for this geomtry 

```python
u_pair = u_source + u_sink
v_pair = v_source + v_sink

width = 10
height = (y_end-y_start)/(x_end-x_start)*width #with this you can control the size of the figure 
pyplot.figure(figsize =(width,height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start,y_end)

pyplot.streamplot(X,Y,u_pair, v_pair, density=2, linewidth=1, arrowsize=2, arrowstyle = '->' )
pyplot.scatter([x_source, x_sink],[y_source, y_sink],color='red', s=80, marker='o')
```

- Define Boundary Condition in the next class

```python
#challlenge 

Ss = -5
x_sink, y_sink =1.0,0.0


ss = 10
x_source, y_source = -1.0,0.0

u_potential = Ss/(2*math.pi)*(np.log((X-x_sink)**2+(Y-y_sink)**2))
u_potentialsink = ss/(2*math.pi)*(np.log((X-x_source)**2+(Y-y_source)**2))

width = 10
height = (y_end-y_start)/(x_end-x_start)*width #with this you can control the size of the figure 
pyplot.figure(figsize =(width,height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start,y_end)

u_p =  u_potential +  u_potentialsink

pyplot.contourf(X, Y, u_p,5)

pyplot.colorbar()

```
